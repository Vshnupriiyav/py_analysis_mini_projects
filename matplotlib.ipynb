```python
import matplotlib.pyplot as plt  
import numpy as np  

# Sample Data: Stock prices over 10 days  
days = np.arange(1, 11)  
prices = [150, 155, 153, 160, 162, 158, 165, 170, 168, 175]  
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Days":days,"prices":prices})
```

```python
df
```

```python
plt.figure(figsize=(8, 5)) 
```

```python
plt.plot(df["Days"],df["prices"],label="stock price")
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("stock prices")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.bar(df["Days"],df["prices"], color=["green","blue","pink"])
plt.show()
```

```python
df = pd.DataFrame({'Height (cm)': np.random.randint(150, 190, 50),  
                   'Weight (kg)': np.random.randint(50, 100, 50)})  

# Scatter plot  
plt.figure(figsize=(8, 5))  
plt.scatter(df['Height (cm)'], df['Weight (kg)'], color='blue', alpha=0.1)  
plt.xlabel('Height (cm)')  
plt.ylabel('Weight (kg)')  
plt.title('Height vs. Weight Distribution')  
plt.grid(True)  
plt.show()  
```

```python
import pandas as pd  

# Creating the DataFrame  
data = {'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),  
        'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A'],  
        'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],  
        'Sales': [500, 300, 450, 700, 600, 750, 650, 400, 520, 580],  
        'Profit': [50, 30, 45, 70, 60, 75, 65, 40, 52, 58]}  

df = pd.DataFrame(data)  

# Display the DataFrame  
print(df)
```

```python
df
```

```python
df["cum_sales"] = np.cumsum(df["Sales"])
```

```python
## Plot a bar chart showing total sales for each product.

plt.bar(df["Product"],df["Sales"])
plt.xlabel("products")
plt.ylabel("sales")
plt.title("Sales by product")
plt.legend()
plt.grid(True) 
plt.show()
```

```python
#Plot a line chart for sales over time.

plt.figure(figsize=(8, 5))  
plt.plot(df['Date'], df['Sales'], marker='o', linestyle='-', color='b')  

plt.xlabel('Date')  
plt.ylabel('Sales')  
plt.title('Sales Over Time')  
plt.xticks(rotation=45)  # Rotate dates for better visibility  
plt.legend(label="sales")  
plt.grid(True)  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np  

# Sample Data  
data = {'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),  
        'Sales': np.cumsum(np.random.randint(50, 150, 10))}  
df = pd.DataFrame(data)  

# Area Plot  
plt.figure(figsize=(8, 5))  
plt.fill_between(df['Date'], df['Sales'], color='skyblue', alpha=0.5)  
plt.plot(df['Date'], df['Sales'], color='blue', marker='o')  
plt.xlabel('Date')  
plt.ylabel('Cumulative Sales')  
plt.title('Cumulative Sales Over Time')  
plt.grid(True)  
plt.xticks(rotation=45)  
plt.show()

```
